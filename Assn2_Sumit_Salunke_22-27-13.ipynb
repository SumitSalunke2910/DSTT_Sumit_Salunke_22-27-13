```python
#Q.4 Retrieve the list of all NaN/Null/Empty cells in the form of Boolean list
#Answer:

import numpy as np
df.replace('--',np.NaN, inplace=True)
df.replace('na',np.NaN, inplace=True)
df.isna()
```

```python
#Q.5 Drop all rows with N/A,NA,na values in Num_Bedrooms
# Answer:
df1=df.dropna(subset='NUM_BEDROOMS')
df1
```

```python
# Q.6 Replace the NaN and String entries in Num_Bath with previous entry
# Answer:
df1.loc[6,'NUM_BATH']=2
df1
```

```python
# Q.7 Replace the empty values in “SQ_FT” with the mean of the all entries.
# Answer:
df1['SQ_FT']= df1['SQ_FT'].astype('float')
mean = df1['SQ_FT'].mean()
df1['SQ_FT'].fillna(mean, inplace=True)
df1
```

```python
# Q.8 What is the sum of SQ_FT of all entries in the dataset.
# Answer:
sum = df1['SQ_FT'].sum()
sum
```

```python
# Q.9 Add a new column “NUM_STORE” and with all the values as 1 for all rows
#Answer:
df1["NUM_STORE"]=1
df1
```

```python
# Q.10 Create a DataFrame containing all entries of TN state only
#Answer:
df1_TN = df1.loc[df1.STATE=='TN',:]
df1_TN
```

```python
# Q.14 Create a DataFrame having SQ_FT area greater than 1000 and display the DataFrame
#Answer:
df1_SQFT = df1.loc[df1.SQ_FT>1000,:]
df1_SQFT
```

```python
# Q.15 Create a DataFrame having only first Three columns and First Three rows
#Answer:
df2= df1.iloc[:3,:3]
df2
```

```python
# Q. 16 Display the state having highest average price per square foot of housing area. Hint: Average of (TotalPrice/Sq_Ft)
#Answer:
df1['AVG_PRICE_PER_SQFT']=df1['PRICE'] / df1['SQ_FT']
df1= df1.sort_values(by='AVG_PRICE_PER_SQFT', ascending=False)
df1.reset_index(inplace=True, drop=True)
df1.loc[0,'STATE']
```
